```python
import json
import os
import pandas as pd
import numpy as np
os.chdir('C:\\Users\\wmj51\\Desktop\\python')

data = []

with open("positive_tweets.json") as f:
    # use 'marid_post_week3 as a test sample'
    for line in f:
        try:
            tweet = json.loads(line)
            if tweet['text'].find('RT @')!=-1:
            # drop Retweeted tweet
                continue
            data.append(tweet)
        except:
            continue

with open('npositive_tweets.json','w') as fp:
    json.dump(data,fp)

tweets_data_path=('npositive_tweets.json')
tweets_data = []
tweets_file = open(tweets_data_path,'r')
tweets_data = json.load(tweets_file)
type(tweets_data)

# convert the json file to dataframe
df = pd.DataFrame(tweets_data)
df.info()
```

```python
col_to_keep = ['text']
df = df[col_to_keep]
df['target'] = 1
df.head()
```

```python
data = []

with open("negative_tweets.json") as f:
    # use 'marid_post_week3 as a test sample'
    for line in f:
        try:
            tweet = json.loads(line)
            if tweet['text'].find('RT @')!=-1:
            # drop Retweeted tweet
                continue
            data.append(tweet)
        except:
            continue

with open('nnegative_tweets.json','w') as fp:
    json.dump(data,fp)

tweets_data_path=('nnegative_tweets.json')
tweets_data = []
tweets_file = open(tweets_data_path,'r')
tweets_data = json.load(tweets_file)
type(tweets_data)

# convert the json file to dataframe
df2 = pd.DataFrame(tweets_data)
df2.info()
```

```python
col_to_keep = ['text']
df2 = df2[col_to_keep]
df2['target'] = 0
df2.head()
```

```python
df_tweets = pd.concat([df, df2])
df_tweets.head()
df_tweets.to_csv('train_NLTKCorpus_Twitter.csv', encoding='utf-8')
```
